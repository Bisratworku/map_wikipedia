```python
import wikipedia
import numpy as np
import json
```

```python
class node:
    def __init__(self, title):
        self.title = title
        self.branches = None
    def grab_links(self):
        page = wikipedia.page(self.title)
        self.branches = [node(i) for i in page.links]
    def iter(self, end):
        self.grab_links()
        def recur(arr , end = 0):
            new = []
            if end <= 0 :
                return
            else:
                for idx,i in enumerate(arr):
                    try:
                        i.grab_links()
                        new.extend(i.branches)
                        print(f'************* {i.title} *********************** {idx}')
                    except:
                        continue
                return recur(new, end - 1)
        recur(self.branches, end)                
```

```python
n = node("Teddy Afro")
```

```python
n.iter(end = 2)
```

```python
def object_to_dict(obj):
    if not hasattr(obj, "__dict__"):
        return obj
    result = {}
    for key, val in obj.__dict__.items():
        if isinstance(val, list):
            result[key] = [object_to_dict(i) for i in val]
        else:
            result[key] = object_to_dict(val)
    return result
```

```python
result = object_to_dict(n)
```

```python
with open("wikipedia.json", 'w') as f:
    json.dump(result, f)
```
