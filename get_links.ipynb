```python
import wikipedia
import numpy as np
```

```python
page = wikipedia.page("Teddy Afro")
```

```python
links = page.links
```

```python
rarr =  []
def arr(link):
    for idx,i in enumerate(link):
        try:
            page = wikipedia.page(i)
            print(f'Appending ***************************************** {page.title} ------------ {idx}')
            rarr.extend(page.links)
        except:
            continue
    return list(set(rarr))
```

```python
u = arr(links)
```
