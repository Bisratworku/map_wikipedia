```python
class link:
    def __init__(self, title:str):
        self.title = title
        self.img = None
        self.summary = ""
        self.branches = None
    def get_links(self):
        page = wikipedia.page(self.title)
        self.img = page.images
        self.branches = [link(i) for  i in page.links]
        self.summary = page.summary
    def move(self, end = 0):
        self.get_links()
        def recur(arr, start = 0,end = 0):
            if end < 0 : 
                return print("ffadaafa")
            else:
                u = []
                for idx, i in enumerate(arr):
                    print(f'__________________________________ {idx} _____________________________________')
                    try:
                        print(i.title)
                        i.get_links()
                        for k in i.branches:
                            u.append(k)
                    except:
                        pass
                return recur(arr = u,end = end - 1)
        recur(self.branches,end = end)
```
